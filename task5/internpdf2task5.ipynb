```python
# List of example support tickets
tickets = [
    "My internet connection is down.",
    "I want to reset my password.",
    "The billing amount is incorrect.",
    "My app crashes on startup.",
    "How can I upgrade my plan?",
    "I can't log into my account.",
    "Please cancel my subscription.",
    "I was overcharged this month.",
    "Where can I change my username?",
    "App is showing black screen."
]
```

```python
# Dictionary of tags and related keywords
tags_keywords = {
    "Billing": ["billing", "payment", "bill", "overcharged", "charged", "invoice"],
    "Technical Issue": ["crash", "down", "error", "bug", "internet", "connection", "black screen"],
    "Login Problem": ["login", "password", "reset", "access", "log into", "authentication"],
    "Upgrade": ["upgrade", "plan", "renew"],
    "Account": ["account", "subscription", "username", "cancel", "profile", "change"]
}
```

```python
def rule_based_tags(text):
    text_l = text.lower()  # convert to lowercase
    scores = {}

    for tag, keywords in tags_keywords.items():
        score = sum(kw in text_l for kw in keywords)  # count matching keywords
        scores[tag] = score

    # Sort tags by match score (descending) and return top 3
    sorted_tags = sorted(scores.items(), key=lambda x: x[1], reverse=True)
    top_tags = [tag for tag, score in sorted_tags if score > 0][:3]
    return top_tags if top_tags else ["Uncategorized"]
```

```python
# Apply tagging function to each ticket
results = []
for ticket in tickets:
    tags = rule_based_tags(ticket)
    results.append({"ticket": ticket, "tags": tags})
```

```python
import pandas as pd

df = pd.DataFrame(results)
df
```
